```python
df['Valuation ($B)'] = (
    df['Valuation ($B)']
    .str.replace('$', '', regex=False)
    .astype(float)
)
```

```python
df['Valuation ($B)'].head()
```

```python
df['Date Joined'] = pd.to_datetime(df['Date Joined'])
```

```python
df['Date Joined'].head()
```

```python
df['Year Joined'] = df['Date Joined'].dt.year
```

```python
df[['Date Joined','Year Joined']].head()
```

Let's start learning groupby(), because it's basically the Pandas equivalent of SQL's GROUP BY.

```python
df.groupby('Country')['Valuation ($B)'].sum().sort_values(ascending=False).head(10)
```

Which industries have the highest total valuation?

```python
df.groupby('Industry')['Valuation ($B)'].sum().sort_values(ascending=False).head(10)
```

```python
df.groupby('Industry')['Valuation ($B)'].agg(
    ['count', 'sum', 'mean']
).sort_values('sum', ascending=False).head(10)
```

Which years produced the most unicorns?

```python
df['Year Joined'].value_counts().sort_index()
```

Which countries have the most unicorns
Which countries have the highest total valuation
Which countries have the highest average valuation

```python
df.groupby('Country')['Valuation ($B)'].agg(
    ['count','sum','mean']
).sort_values('sum', ascending=False).head(10)
```
